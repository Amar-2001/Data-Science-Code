```python
from flask import Flask,render_template, request, jsonify

app=Flask(__name__)


#html aur css page se connect kiye
@app.route("/",methods=["GET","POST"]) # route(addresss) de diye aur method bhi 
                                     # laga diye get aur post
def home_page():
    return render_template("index.html")# return kar diye render template
                                        #  ke andar index.html ko

@app.route("/math",methods=["POST"]) # post method yaani body ke through kar rahe na operation
                                  # is ka naame math rakhe hai
def math_operation():
    if (request.methods=="POST"): # request ka method agar post hai yaani body ke through
                                #pehle isme mein likhte thay request.args.get par ab ye 
        ops=request.form['operation'] # request kiye form ko jisme operation ho raha(operation html pages mei hai)
        num1 = int(request.form['num1']) #request kiye form ko jisme num1 hai
                                        #(num1 html mein hai) aur usko int mein 
                                        # typecast kar diye kyunki pehle wo string mein tha
                                        #  html mein 
        num2 = int(request.form['num2'])#request kiye form ko jisme num2 hai(num1 html mein hai)

        if (ops=='add'):
            r=num1+num2
            result='The sum of '+str(num1) + " and " + str(num2) + " is " +str(r)

        if (ops=='subtract'):
            r=num1-num2
            result='The difference of '+str(num1) + " and " + str(num2) + " is " +str(r)

        if (ops=='multiply'):
            r=num1*num2
            result='The multiplication of '+str(num1) + " and " + str(num2) + " is " +str(r)

        if (ops=='divide'):
            r=num1/num2
            result='The division of '+str(num1) + " and " + str(num2) + " is " +str(r)

            return render_template("result.html",result=result)# return kar do 
            # render template ko (result.html )ko aur uske result mein result ki value 
            # daal do 




# again

@app.route("/postman_data",methods=["POST"]) # post method yaani body ke through kar rahe na operation
                                  # is ka naame math rakhe hai
def math_operation1():
    if (request.methods=="POST"): # request ka method agar post hai yaani body ke through
                                #pehle isme mein likhte thay request.args.get par ab ye 
        
        # sirf request.form ke jagah request.json kar diye OK
        ops=request.json['operation'] # request kiye form ko jisme operation ho raha(operation html pages mei hai)
        num1 = int(request.json['num1']) #request kiye form ko jisme num1 hai
                                        #(num1 html mein hai) aur usko int mein 
                                        # typecast kar diye kyunki pehle wo string mein tha
                                        #  html mein 
        num2 = int(request.json['num2'])#request kiye form ko jisme num2 hai(num1 html mein hai)

        if (ops=='add'):
            r=num1+num2
            result='The sum of '+str(num1) + " and " + str(num2) + " is " +str(r)

        if (ops=='subtract'):
            r=num1-num2
            result='The difference of '+str(num1) + " and " + str(num2) + " is " +str(r)

        if (ops=='multiply'):
            r=num1*num2
            result='The multiplication of '+str(num1) + " and " + str(num2) + " is " +str(r)

        if (ops=='divide'):
            r=num1/num2
            result='The division of '+str(num1) + " and " + str(num2) + " is " +str(r)


            return jsonify(result)

            # neeche return ki koi jarurat nahi aur waise bhi ek function mein ek hi return kaam karta
            # return render_template("result.html",result=result)# return kar do 
            # # render template ko (result.html )ko aur uske result mein result ki value 
            # # daal do 


if __name__ =="__main__":
    app.run(host="0.0.0.0")
```
