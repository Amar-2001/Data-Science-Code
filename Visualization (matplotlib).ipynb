```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10, 200) # 0 se 10 tak 200 data
x
```

```python
y=np.sin(x)
y
```

```python
plt.plot(x,y) # looks like sin graph because hum ne sin mein pass kiya value
```

```python
y1=np.cos(x)
plt.plot(x,y1)
```

```python
y2=np.tan(x)
plt.plot(x,y2)
```

```python
plt.plot(x,y)
plt.xlabel("This is my x label")
plt.ylabel("This is my y label")
plt.title("This is my title")
plt.show()
```

```python
x2=np.random.rand(50)
y2=np.random.rand(50)
colour=np.random.rand(50)
plt.figure(figsize=(8,4))
plt.scatter(x2,y2, c=colour)
# plt.grid()
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter graph")
plt.show()
```
